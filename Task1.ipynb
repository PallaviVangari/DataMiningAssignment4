```python
import pandas as pd

# Load the data
file_path = '/content/sample_data/shopping_trends.csv'  # Adjust the path if needed
data = pd.read_csv(file_path)

# Display basic information about the dataset
print(data.info())
print(data.head())

# Handling missing values (example: dropping or imputing)
# data.dropna(inplace=True)  # To drop rows with missing values
# data.fillna(method='ffill', inplace=True)  # To fill missing values

# Descriptive statistics
print(data.describe(include='all'))

# For categorical data, count the occurrences
print(data['Category'].value_counts())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram for a numerical column (e.g., "Purchase Amount (USD)")
plt.figure(figsize=(10, 6))
sns.histplot(data['Purchase Amount (USD)'], kde=True)
plt.title('Distribution of Purchase Amounts')
plt.xlabel('Purchase Amount (USD)')
plt.ylabel('Frequency')
plt.show()

# Box plot for the same numerical column
plt.figure(figsize=(10, 6))
sns.boxplot(x=data['Purchase Amount (USD)'])
plt.title('Box Plot of Purchase Amounts')
plt.show()
```

```python
# Bar chart for a categorical column (e.g., "Gender")
plt.figure(figsize=(10, 6))
sns.countplot(x=data['Gender'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

# Pie chart for the same categorical column
gender_counts = data['Gender'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Gender Distribution')
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Grouped bar chart for average purchase amount by gender
grouped_data = data.groupby('Gender')['Purchase Amount (USD)'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Gender', y='Purchase Amount (USD)', data=grouped_data)
plt.title('Average Purchase Amount by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Purchase Amount (USD)')
plt.show()
```
